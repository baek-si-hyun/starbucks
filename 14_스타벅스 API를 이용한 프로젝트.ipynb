```python
import warnings
warnings.filterwarnings('ignore')
import json
import requests
from pandas.io.json import json_normalize
import folium
```

```python
targetSite = 'https://www.starbucks.co.kr/store/getSidoList.do'
request = requests.post(targetSite, data={
    'rndCod':'UXR15FAQFM'
})

print(request) # <Response [200]>
print(type(request))
print(request.text)
```

```python
star_json = json.loads(request.text)
print(type(star_json))
print(star_json)
```

```python
sidoList = request.json()
print(type(sidoList))
print(sidoList)
```

```python
print(len(sidoList))
print(len(sidoList['list']))
print(len(sidoList.get('list')))
```

```python
sido_list = dict()
for sido in sidoList['list']:
  # print('{} : {}'.format(sido['sido_cd'], sido['sido_nm']))
  sido_list[sido['sido_cd']] = sido['sido_nm']
print(sido_list)
```

```python
sido_cd = input('구군 코드를 검색할 시도 코드를 입력하세요: ')
targetSite = 'https://www.starbucks.co.kr/store/getGugunList.do'
request = requests.post(targetSite, data={
    'sido_cd': sido_cd,
    'rndCod': 'O1LIU3XBI3'
})

# print(request)
# print(type(request.text))
gugunList = request.json()
# print(gugunList['list'][0])
# print(type(gugunList))

gugun_list = dict()
for gugun in gugunList['list']:
  gugun_list[gugun['gugun_cd']] = gugun['gugun_nm']
print(gugun_list)
```

```python

targetSite = 'https://www.starbucks.co.kr/store/getStore.do?r=KO1XF3AWAI'
request = requests.post(targetSite, data={
    'ins_lat': 37.5108295,
    'ins_lng': 127.02928809999999,
    'p_sido_cd': '01',
    'p_gugun_cd': '',
    'in_biz_cd': '',
    'iend': 2000, # 서버가 응답하는 최대 매장의 갯수
    'set_date': ''
})

star_json= request.json()
print(len(star_json['list']))
print(type(star_json))

star_df = json_normalize(star_json, 'list')
print(type(star_df))
```

```python
star_df_map = star_df[['s_name', 'sido_code', 'sido_name', 'gugun_code', 'gugun_name', 'doro_address', 'lat', 'lot']]
star_df_map
```

```python
star_df_map.dtypes
```

```python
star_df_map['lat'] = star_df_map['lat'].astype(float)
star_df_map['lot'] = star_df_map['lot'].astype(float)
star_df_map.dtypes
```

```python
star_map = folium.Map(location=[37.49972, 127.034908], zoom_start=18)

popup = folium.Popup('아크플레이스', max_width=200)
folium.Marker(location=[37.49972, 127.034908], popup=popup).add_to(star_map)

popup = folium.Popup('역삼포스코', max_width=200)
folium.Marker(location=[37.499721, 127.033582], popup=popup).add_to(star_map)

star_map.save('./star1.html')
star_map
```

```python
# 서울 특별시에 위치한 스타벅스 전 지점을 지도위에 표기
star_map = folium.Map(location=[star_df_map['lat'].mean(), star_df_map['lot'].mean()], zoom_start=14)

for index, data in star_df_map.iterrows():
  popup = folium.Popup(data['s_name'] + '점, 주소: ' + data['doro_address'], max_width=200)
  folium.Marker(location=[data['lat'], data['lot']], popup=popup).add_to(star_map)

star_map.save('./star1.html')
star_map
```
